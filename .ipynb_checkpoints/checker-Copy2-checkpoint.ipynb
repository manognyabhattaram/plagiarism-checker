```python
import nltk
from nltk.corpus import stopwords
from nltk.tokenize import word_tokenize
from nltk.stem import PorterStemmer
from nltk.stem import WordNetLemmatizer
nltk.download('stopwords')
nltk.download('punkt')
nltk.download('wordnet')


from os.path import isfile
from os.path import join

import os
from num2words import num2words
import numpy as np
import string
import pandas as pd
import copy
import pickle
import re
import math
```

```python
def lowercase(data):
    return np.char.lower(data)
```

```python
def remove_stopwords(data):
    stop_words = stopwords.words('english')
    words = word_tokenize(str(data))
    new = ""
    for word in words:
        if word not in stop_words and len(word) > 1:
            new = new + " " + word
    return new
```

```python
def remove_punct(data):
    punct = "!\"#$%&()*+-./:;<=>?@[\]^_`{|}~\n"
    for i in range(len(punct)):
        data = np.char.replace(data, punct[i], ' ')
        data = np.char.replace(data, " ", " ")
    data = np.char.replace(data, ',', '')
    return data
```

```python
def remove_apostrophes(data):
    data = np.char.replace(data, "'", "")
    data = np.char.replace(data, "â\x80\x98", "")
    data = np.char.replace(data, "â\x80\x99", "")
    return data
```

```python
def stemming(data):
    stemmer = PorterStemmer()
    tokens = word_tokenize(str(data))
    new = ""
    for word in tokens:
        new = new + " " + stemmer.stem(word)
    return new
```

```python
def lemmatize(data):
    lemmatizer = WordNetLemmatizer()
    tokens = word_tokenize(str(data))
    new = ""
    for word in tokens:
        new = new + " " + lemmatizer.lemmatize(word)
    return new
```

```python
def preprocess(data):
    data = lowercase(data)
    data = remove_punct(data)
    data = remove_apostrophes(data)
    data = remove_stopwords(data)
    data = num_to_words(data)
    data = lemmatize(data)
    data = stemming(data)
    data = remove_punct(data)
    data = num_to_words(data)
    data = lemmatize(data)
    data = stemming(data)
    data = remove_punct(data) #done again to remove hyphens produced by num2words
    data = remove_stopwords(data) #done agan to remove stopwords produced by num2words
    return data
```

```python
processed = []
path = "../texts/"
files = [document for document in os.listdir(path) if document.endswith('.txt')]
#file_contents = [open(os.path.join(path, file), 'r', encoding = "unicode_escape").read() for file in files]
```

```python
for file in files:
    file.encode('utf8').strip() #encodes each of the files into utf-8
    fh = open(os.path.join(path, file), 'r', encoding = "utf-8")
    file_content = fh.read()

    processed.append(word_tokenize(str(preprocess(file_content))))
```

```python
#calculating Document Frequency for all words

df = {}

for i in range(len(processed)):
    tokens = processed[i]
    for token in tokens:
        try:
            df[token].add(i)
        except:
            df[token] = {i}

for i in df:
    df[i] = len(df[i])
```
